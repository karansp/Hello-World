```python
loan_data['Gender'] = loan_data['Gender'].astype(str)
loan_data['Gender'] = loan_data['Gender'].replace('Male','1')
loan_data['Gender'] = loan_data['Gender'].replace('Female','0')
loan_data.head()
```

```python
loan_data['Self_Employed'] = loan_data['Self_Employed'].astype(str)
loan_data['Self_Employed'] = loan_data['Self_Employed'].replace('Yes','1')
loan_data['Self_Employed'] = loan_data['Self_Employed'].replace('No','0')
loan_data.head()
```
